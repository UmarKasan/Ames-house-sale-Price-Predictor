# Realtor webscrape

Import Python packages

```python
import requests,pandas,math,re, random, time
from bs4 import BeautifulSoup
```

```python
session=requests.Session()
session.headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
url='https://www.realtor.com'
re_list=[]
realtor_default_count=44
page_counter=1
search_url=''.join([url, '/realestateandhomes-search/Ames_IA'])
page=session.get(search_url)
c=page.content
soup=BeautifulSoup(c,"html.parser")
```

```python
print(soup)
```

```python
total_homes=soup.find("section",{"id":"srp-footer-found-listing"}).text.strip().replace("\n","")
print(total_homes)
```

```python
total_homes=int(re.sub('[^0-9]','', total_homes))
total_pages=math.ceil(total_homes/realtor_default_count)
```

```python
while page_counter <= total_pages:
    print(page_counter)
    if page_counter!=1:
        search_url=''.join([url, '/realestateandhomes-search/Ames_IA/pg-',(str(page_counter))])
        print(search_url)
        page=session.get(search_url)
        c=page.content
        soup=BeautifulSoup(c,"html.parser")
    re_data=soup.find_all("li", {"class":"jsx-3900997218 component_property-card"})
    for item in re_data:
        re_dict={}
        re_dict["Property Type"]=item.find("span",{"class":"jsx-1547078465"}).text
    
        try:
            re_dict["Price"]=item.find("span",{"data-label":"pc-price"}).text
        except:
            re_dict["Price"]="No Value"
        try:
            community = item.find("div",{"data-label":"pc-address"}).text.strip()
            community = community + " "
        except:
            pass
        try:
            re_dict["Address"]=community + item.find("div",{"data-label":"pc-address"}).text.strip().replace(",","")
        except:
            re_dict["Address"]="No Address"
        try:
            re_dict["Locality"]=item.find("div",{"data-label":"pc-address-second"}).text + ", " + item.find("span",{"class":"listing-region"}).text + " " + item.find("span",{"class":"listing-postal"}).text
        except:
            re_dict["Locality"]="No Value"
        try:
            re_dict["Beds"]=item.find("li",{"data-label":"pc-meta-beds"}).text.strip()
        except:
            re_dict["Beds"]="No Value"
        try:
            re_dict["Baths"]=item.find("li",{"data-label":"pc-meta-baths"}).text.strip()
        except:
            re_dict["Baths"]="No Value"
        try:
            re_dict["Home Size"]=item.find("li",{"data-label":"pc-meta-sqft"}).text.strip()
        except:
            re_dict["Home Size"]="No Value"
        try:
            re_dict["Lot Size"]=item.find("li",{"data-label":"pc-meta-sqftlot"}).text.strip()
        except:
            re_dict["Lot Size"]="No Value"
        try:
            re_dict["Garage"]=item.find("li",{"data-label":"pc-meta-garage"}).text.strip()
        except:
            re_dict["Garage"]="No Value"
        try:
            re_dict["Construction"]=item.find("span",{"data-label":"pc-label-new_construction"}).text.strip()
        except:
            pass
#         re_dict["Link"]=url + item['url']
        re_list.append(re_dict)
        community=""
        sleep_duration = random.randint(1,2)
        print(sleep_duration)
        time.sleep(sleep_duration)
    page_counter+=1
```

```python
print(re_list)
```

```python
re_df=pandas.DataFrame(re_list)
re_df.to_csv("./datasets/RealtorDataAmes.csv")
```
